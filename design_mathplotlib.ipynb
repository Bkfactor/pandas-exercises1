```python
import matplotlib.pyplot as plt
import numpy as npy
```

```python
x=npy.linspace(10,2,11)
y= x**2
x
```

### Figure Size and DPI

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,8), dpi=100)
#axx = fig1.add_axes([0,0,1,1])
axes[0].plot(x,x**2, 'b', label='x square')
axes[0].plot(x,x**3,'r', label='x cube')
axes[0].set_title('fig1')
axes[0].set_xlabel('xaxis')
axes[0].set_ylabel('yaxis')


axes[1].plot(y,x)
axes[1].set_title('fig2')
axes[1].set_xlabel('xaxis')
axes[1].set_ylabel('yaxis')
plt.tight_layout()

axes[0].legend(loc=0)
```

```python
fig.savefig("my_picmath2.png",dpi=100)
```

### PLOT APPERANCE

```python
fig= plt.figure(figsize=(8,3))

ax= fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'blue',lw=2,alpha=0.8, ls='--',
        #marker='o',markersize=10,markerfacecolor='grey',
        #markeredgewidth=2, markeredgecolor='green'
       )#lw means linewidth,ls mean linestyle
ax.set_xlim(0,10)
ax.set_ylim(40,100)
```
