```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Start Finish in points
```

```python
t= np.linspace(0,40,1000)
```

```python
#Distance travelled by robber d=2.5t
```

```python
d_robber= 2.5 * t
```

```python
#Distance travelled by Sheriff d=3(t-5)
```

```python
d_sherrif = 3*(t - 5)
```

```python
fig, ax = plt.subplots()
plt.title('How Robber Was Caught')
plt.xlabel('Time Taken in Min')
plt.ylabel('Distance Covered in km')
ax.set_xlim(0,40)
ax.set_ylim(0,100)
ax.plot(t,d_robber, c='red')
ax.plot(t,d_sherrif,c='blue')
plt.axvline(x=30,color='green',ls='--')
plt.axhline(y=75,color='green',ls='--')
```

```python

day= np.linspace(0,60,600)
```

```python
a_design=day * 1
```

```python
b_design=  4 * (day-30)
```

```python
fig,axe=plt.subplots()
plt.title('Energy Production')
plt.xlabel('Days')
plt.ylabel('Energy produce in kj')
axe.set_xlim(0,60)
axe.set_ylim(0,100)
axe.plot(day,a_design,color='blue')
axe.plot(day,b_design,color='purple')
plt.axvline(x=40,color='green',ls='--')
plt.axhline(y=40,color='green',ls='--')
```
