```python
import seaborn as sns
%matplotlib inline 
import matplotlib.pyplot as mpl
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Distribution
```

```python
distill= sns.displot(tips['total_bill'],kde=True,bins=100)
distill
mpl.show()
```

### Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
mpl.show()
```

```python
##Pairplot
```

```python
sns.pairplot(tips,hue ='sex',palette='coolwarm')
mpl.show()
```

```python
## rug plot
```

```python
sns.rugplot(tips['total_bill'])
mpl.show()
```

```python
sns.kdeplot(tips['total_bill'])
mpl.show()
```

## Categorical Data

```python
import numpy as ny
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=ny.std)
mpl.show()
```

```python
sns.countplot(x='sex',data=tips)
mpl.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
mpl.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
mpl.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex' ,dodge=True)
mpl.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
mpl.show()
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
mpl.show()
#general plot call
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
tips.head()
```

```python
hm=tips.corr(numeric_only=True)
```

```python
sns.heatmap(hm, annot=True,cmap='coolwarm')
mpl.show()
```

```python
flights
```

```python
fphc=flights.pivot_table(index='month',columns='year', values='passengers', observed=False)
```

```python
sns.heatmap(fphc,cmap='coolwarm',linecolor='black',linewidth=2)
mpl.show()
mpl.tight_layout
```

```python
sns.clustermap(fphc,cmap='coolwarm',standard_scale=1,linecolor='black',linewidth=2)
mpl.show()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

## Grids

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
mpl.show()
```

```python
grid=sns.PairGrid(iris)
grid.map_diag(sns.histplot)
grid.map_upper(mpl.scatter)
grid.map_lower(sns.kdeplot)
mpl.show()
```
