```python
import matplotlib.pyplot as plt
import numpy as ny
```

```python
%matplotlib inline
```

```python
x = ny.linspace(0,20,7)
y= x**2
z=x+2
```

```python
#functional method
plt.plot(z,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('First-Plot')
```

```python
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g')

plt.subplot(1,2,2)
plt.plot(y,z, 'y')
plt.show()

plt.subplot(1,2,2)
plt.plot(y,x, 'r')
plt.show()
```

```python

#ObjectOriented
fig = plt.figure()

axxis = fig.add_axes([0.1,0.1,0.8,0.8])
axxis.plot(x,y)
axxis.set_xlabel('x-axis')
axxis.set_ylabel('y-axis')
plt.show()
```

```python
fig=plt.figure()
a1= fig.add_axes([0.1,0.1,0.8,0.8])
a2=fig.add_axes([0.2,0.45,0.4,0.4])
a1.plot(x,y)
a1.set_title('Main Plot')
a2.plot(y,x)
a2.set_title('Small Plot')
plt.show()
```
